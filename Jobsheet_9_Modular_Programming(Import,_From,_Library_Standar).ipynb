```python
import konversi

def celcius_to_fahrenheit(celsius):
  fahrenheit = (celsius * 9/5) + 32
  return fahrenheit

def fahrenheit_to_celcius(fahrenheit):
  celsius = (fahrenheit - 32) * 5/9
  return celsius

celsius_temp = 25
fahrenheit_temp = celcius_to_fahrenheit(celsius_temp)
print(f"{celsius_temp} Celcius sama dengan {fahrenheit_temp:.2f} Fahrenheit")

fahrenheit_temp = 77
celsius_temp = fahrenheit_to_celcius(fahrenheit_temp)
print(f"{fahrenheit_temp} Fahrenheit sama dengan {celsius_temp:.2f} Celcius")
```

#Tugas No 08: Studi Kasus Mini

```python
%%writefile laporan.py

def tampilkan_skor(total_skor, jumlah_soal):
  """Menampilkan ringkasan skor kuis."""
  print(f"\n--- Ringkasan Skor ---")
  print(f"Anda memperoleh skor {total_skor} dari {jumlah_soal} soal.")
  print(f"Persentase skor: {(total_skor / jumlah_soal) * 100:.2f}%")
```

```python
import random
import laporan

jumlah_soal = 5
total_skor = 0

for i in range(jumlah_soal):
  angka1 = random.randint(1, 20)
  angka2 = random.randint(1, 20)

  print(f"\nSoal ke {i + 1}: Berapa hasil dari {angka1} + {angka2}?")
  jawaban_pengguna = int(input("Jawaban Anda: "))

  jawaban_benar = angka1 + angka2
  if jawaban_pengguna == jawaban_benar:
    print("Jawaban Anda benar!")
    total_skor += 1
  else:
    print(f"Jawaban Anda salah. Jawaban yang benar adalah {jawaban_benar}.")

laporan.tampilkan_skor(total_skor, jumlah_soal)
```
